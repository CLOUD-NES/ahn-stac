```python
from rustac import DuckdbClient
import geopandas as gpd
```

# Searching STAC AHN collections

Search one of the [AHN stac-geoparquet collections](https://github.com/CLOUD-NES/ahn-stac):

```python
client = DuckdbClient()
table = client.search_to_arrow(
    "https://raw.githubusercontent.com/cloud-nes/ahn-stac/main/data/AHN4.parquet",
    bbox=[5.085297, 52.050390, 5.197220, 52.117516],
)
```

Load items using geopandas:

```python
items = gpd.GeoDataFrame.from_arrow(table)
items = items.set_crs("EPSG:4326")
items
```

Visualize search results on a base map:

```python
items.explore()
```

The metadata contains links to the actual datasets:

```python
items["assets"][0]
```
